```python
#6 extract multiple columns from existing columns
# Date -day, month, year, weekend, Quarter
```

```python
odi['MatchDate'] = pd.to_datetime(odi['MatchDate'])
odi.info()
```

```python
odi['day'] = odi['MatchDate'].dt.day
odi['month'] = odi['MatchDate'].dt.month
odi['year'] = odi['MatchDate'].dt.year
odi['quarter'] = odi['MatchDate'].dt.quarter
odi['weekend'] = odi['MatchDate'].dt.day_name().isin(['Saturday','Sunday'])
odi
```

```python
odi.shape
```

```python
odi.describe(include='datetime')
```

```python
odi['Runs'].head()
```

```python

#Runs- Century, Fifty, Duckouts, Missed Century,type_of_run



odi['Century']=odi['Runs'].apply(lambda x: 1 if x>=100 else 0)



odi['Fifty']=odi['Runs'].apply(lambda x: 1 if x>=50 and x<=99 else 0)


odi['Duckouts']=odi['Runs'].apply(lambda x: 1 if x==0 else 0)


odi['Missed_Century']=odi['Runs'].apply(lambda x: 1 if x>=95 and x<100 else 0)

```

### For every player from India calculate the season-wise total runs scored between 2006 and 2011


```python
odi
```

```python
odi_india = odi[(odi.year>=2006) & (odi.year<=2011)&(odi.Country=='India')]
odi_india
```

```python
pt = odi_india.pivot_table(index='Player', values='Runs',columns= 'quarter',
              aggfunc=sum, fill_value=0,margins=True).sort_values(by="All", ascending=False).head(10)[1:]
pt
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(pt.drop('All', axis=1), cmap="YlGnBu",annot=True, fmt = '.4g', cbar=True)
```

```python
dfPivot = odi_india.pivot_table(index='Player', values='Century', columns='year',
                                aggfunc='sum', fill_value=0, margins=True).sort_values(by='All', ascending=False).head(10)[1:]
dfPivot
```

```python
plt.figure(figsize=(10, 8))
dfPivot.drop('All', axis=1).plot.bar(stacked=True)
```

### Binning or Bucketing
- groups continous data into intervals

```python
year_bins=[1990,2000,2005]
```

```python
np.digitize([1989,1990,2000,1975,2004,1995,2005,2011],bins=year_bins)
```

```python
labels={0:'1974-1989',
        1:'1990-1999',
        2:'2000-2004',
        3:'2005 onwards'}
```

```python
odi['year_bin']=odi['year'].apply(lambda v:np.digitize(v,bins=year_bins)).replace(labels)
odi['year_bin'].value_counts()
#odi
```
